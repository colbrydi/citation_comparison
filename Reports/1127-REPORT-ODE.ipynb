Innovation diffusion is a theory that seeks to explain how new ideas and technologies spread through cultures. The rise of social media has provided a new platform to study the diffusion of innovation. I am quite interested in the topic of innovation diffusion across online social networks. But I haven't had a chance to consider how to use ordinary differential equation (ODE) models for this classical problem. After reading several literatures related to social network modeling with ODEs, I summarize the following contents. 


Firstly, let's define the **Innovation diffusion over online social network**. It refers to **the process by which a piece of innovation spreads and reaches individuals through interactions in a network**[1]. An innovation, for instance, can be an idea, practice, a piece of news being reported many times across social networks. Based on the order of adoption time, it has been found that there are five types of adopters: 1) Innovators(top 2.5%), 2) Early adopters(13.5%), 3) Early majority(34%), 4) Late majority(34%), and 5) Laggards(16%) [2]. The figure below shows the distribution of the five categories of adopters as well as the cumulative adoption **S-shaped curve**.

[Note]:

- *The S-shaped (sigmoid) curve (Description)*: it indicates that the adoption rate is relatively slow when innovators or early adopters begin to adopt the product or information. As soon as the majority of early adopters are onboard, the adoption curve becomes linear, and the rate is close to constant until all late majority members begin to adopt the product. After the late majority adopts the product, the adoption rate gradually becomes slow once again when laggards start adopting. In the end, the curve slowly approaches 100% [1].

- *The S-shaped (sigmoid) curve (Type)*: in general, most innovation patterns follow a generally S-shaped curve. But the exact form of each curve, such as the slope and the asymptote, can be different depending on specific information or diffusion process. 
    
    


Second, **mathematical models such as ODEs can represent/model the S-shaped growth curve**, or the spread of innovation among a set of prospective adopters across social networks over time. For example, Mahajan [3] presents several ODEs to study the observed diffusion patterns based on the assumption that prespecified trends or distribution functions are given. The **basic innovation diffusion model** takes the following forms:

$$\frac{dN(t)}{dt} = g(N,t)[\bar{N} - N(t)]$$

where the initial condition $N(0) = N_0$, $N(t)$ is the cumulative number of adopters at time $t$ in a social network. $\bar{N}$ is the total number of potential adopters. $g(N,t)$ denotes the coefficient of diffusion, which is a function of current number of adopter $N$ and time $t$. 

[Note]:
- *This equation states* that the rate of diffusion of innovation at any time $t$ is directly proportional to the difference between the total number of potential adopters and the number of previous adopters at that time. 

- *This equation has a closed-form solution N(t)*, which is the cumulative number of prior adopters in given social networks.

- *The coefficient of diffusion, $g(N,t)$, can have different forms* that might be dependent on many characteristics of the diffusion process, such as communication channels and innovation types.
    - 1)the External-influence diffusion model: when $g(N,t) = \alpha$ (the coefficient of diffusion is independent of the current number of adopters)
    
    - 2)the Internal-influence diffusion model: when $g(N,t) = \beta*N(t)$ (the coefficient of diffusion is linearly proportional to the current number of adopters)
    
    - 3)the Mixed-influence diffusion model: when $g(N,t) = \alpha + \beta*N(t)$(\alpha is a parameter to represent external influence and \beta for internal influence)
