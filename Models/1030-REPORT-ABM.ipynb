# <center>Applications with Agent Based Models </center>

<center>by Jiaoping Chen</center>

Agent-based models (ABM) generated from the Ant-Colony optimization are one of the computational models by simulating specific agents' actions to investigate their impacts on the system. I have taken ABM as one of simplified "multi-agent systems" or "multi-agent simulations" for granted. Recently, I noticed their slight differences from Wikipedia, "*Agent-based modeling is related to, but distinct from, the concept of multi-agent systems in that **the goal of ABM is to search for explanatory insight into the collective behavior of agents obeying simple rules**, typically in natural systems, rather than in designing agents or solving specific practical or engineering problems*"[1].

Most researchers utilized ABM to optimize search routines with a minimum time. Before searching for ABM real-world applications, I can imagine how many resources online are related to optimizing delivery or car-sharing routines, such as FedEx or UberPool. Because they can reduce costs and increase profits. However, two papers caught my eyes, which hit me so hard about how meaningful this ABM model is. Perez et al(2018)[2] investigated how to operate a fleet of unmanned air vehicles (UAVs) to look for a lost target in the minimum possible time, to determine the trajectories. The other paper is from Zheng et al(2019)[3]. In this paper, they collaborated human and UAV together to solve the search-planning problem, to minimize the expected time for human rescuers to reach the target. Because their problems were highly complex, they proposed a bioinspired algorithm to solve the problem instead of the relatively simple ABM. But the algorithm and ABM are both based on an evolutionary algorithm, to simulate all potential actions. One more take-away from this article is **Figure 1** below. It summarizes recent literature about UAV path planning methods, which helps me generate a big picture of how various algorithms can be used to solve this meaningful problem.

I only found a few articles that investigate individuals' behaviors in social networks using ABM. Weng et al(2012)[4] is one of the most interesting articles. The main idea is to use a parsimonious ABM to study THE COMPETITION AMONG IDEAS/MEMES ON SOCIAL MEDIA. That is, whether competition among ideas raised from social media, may affect the popularity of different memes, the diversity of information we are exposed to, and the fading of our collective interests for a specific topic. Here, they smartly define the **Agent** as an individual who shares msg on a social network, but only pays attention to a portion of the information they receive. Specifically, like ants keep search foods in one area, users on a social network keep transmitting information back and forth, such as posts/tweets/retweets on Twitter. Besides, those retweets with a specific hashtag can carry a meme from user to user. As a meme spreads in this way, it forms a cascade or diffusion network. Based on this scenario, this paper further makes a similar assumption like fading smell patterns in ant forging problems. That is the role of **the limited attention of individual users in the information diffusion process**. For example, a user might retweet #basketball on day 1, then we assume the fading attention on basketball if he/she does not retweet similar sports hashtags. 

Weng et al(2012)[4] finally concludes that massive heterogeneity in the popularity and persistency of memes(virality) of different msg on Twitter can be explained by only two factors: a) competition for our limited attention; b) the structure of the social network. Other factors, such as the appeal of the msg content, the persuasiveness of an individual, and external events, were not necessary to derive the obs empirical patterns on Twitter. 

Another recent paper from Iyer et al (2015)[5] found that competition for attention also explains the fact that a small proportion of individuals is responsible for most of the information shared on the social network.
